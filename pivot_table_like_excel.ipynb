```python
import pandas as pd
```

```python
data = [
    ["时间", "数量", "金额", "成本", "税率"],
    ["2020/1/1", 44, 48, 28, 0.01],
    ["2020/1/1", 96, 15, 36, 0.03],
    ["2020/2/1", 36, 84, 45, 0.09],
    ["2020/2/1", 99, 65, 70, 0.09],
    ["2020/3/1", 65, 15, 35, 0.16],
    ["2020/3/1", 11, 32, 18, 0.13],
    ["2020/4/1", 69, 94, 48, 0.13],
    ["2020/4/1", 100, 64, 61, 0.13],
    ["2020/5/1", 80, 86, 53, 0.03],
    ["2020/5/1", 84, 75, 59, 0.16],
    ["2020/6/1", 27, 51, 89, 0.16],
    ["2020/6/1", 47, 83, 91, 0.16],
]
```

```python
# 创建 DataFrame
df = pd.DataFrame(data[1:], columns=data[0])
```

```python
df
```

```python
# 使用 pivot_table 方法将 '税率' 设为列索引，'时间' 设为行索引
pivot_df = df.pivot_table(values=['数量', '金额', '成本'], index='时间', columns='税率', aggfunc='sum', fill_value='').reset_index()
```

```python
pivot_df.sort_index(axis=1, level=1).set_index('时间')
```
